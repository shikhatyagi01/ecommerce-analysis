```python
#import mysql.connector
from mysql import connector
from dotenv import load_dotenv
import pandas as pd
```

```python
import os
load_dotenv()

user= os.getenv("USER")
password= os.getenv('PASSWORD')
host= os.getenv("HOST")

database="swiftmarket"
```

```python
connection=connector.connect(
    user= user,
    password= password,
    host= host,
    database= database

)
cursor = connection.cursor()
```

```python
connection.is_connected()
```

```python
query ='''show tables'''
cursor.execute(query)
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# dimension tables: all tables containing primary keys and connected to fact table
# fact table: e.g sales data

query= '''select *from salesdata limit 5'''
cursor.execute(query)
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
def fetch_salesdata(cursor):
    query = '''SELECT * FROM salesdata LIMIT 5'''
    cursor.execute(query)
    read_query = cursor.fetchall()
    return read_query



read_query = fetch_salesdata(cursor)

for row in read_query:
    print(read_query)
```

```python
def read_query(query):
    cursor.execute(query)
    rows=cursor.fetchall()
    return pd.DataFrame(data=rows,columns=cursor.column_names)
```

```python
query='''select *from customers limit 5'''
read_query(query)
```

```python
query='''select s.saledate, c.firstname, s.totalprice
from salesdata as s
inner join customers as c on
c.Customerid = s.customerid
order by s.totalprice
limit 10'''

read_query(query)
```
